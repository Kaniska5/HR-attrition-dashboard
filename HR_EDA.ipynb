```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('IBM_HR.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns=['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'])
print("Cleaned shape:", df.shape)
```

```python
attrition_counts = df['Attrition'].value_counts()
print(attrition_counts)

plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title('Attrition Distribution')
plt.show()
```

```python
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print(attrition_rate)
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Department', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Department')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20, palette='Set2')
plt.title('Attrition by Age')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Overtime')
plt.show()
```

```python
plt.figure(figsize=(16,10))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set2')
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='MaritalStatus', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Marital Status')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y='JobRole', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Job Role')
plt.show()
```

```python
print("=== EDA Summary ===")
print(f"Total Employees: {len(df)}")
print(f"Attrition Rate: {round(df['Attrition'].value_counts(normalize=True)['Yes']*100, 2)}%")
print(f"Avg Age of employees who left: {round(df[df['Attrition']=='Yes']['Age'].mean(), 1)}")
print(f"Avg Monthly Income of employees who left: {round(df[df['Attrition']=='Yes']['MonthlyIncome'].mean(), 2)}")
print(f"Overtime & Attrition - Yes/Yes count: {len(df[(df['OverTime']=='Yes') & (df['Attrition']=='Yes')])}")
```
